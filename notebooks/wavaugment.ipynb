```python
audio_path = "/content/drive/MyDrive/nguyenanh-projects/streaming-asr/audio_exp/viet/viet_sample.wav"
```

```python
import torchaudio

audio_array, rate = torchaudio.load(audio_path)
```

```python
import augment
effect_chain = augment.EffectChain().pitch(100).rate(16_000)
```

```python
import numpy as np
```

## Pitch shift ⭐⭐⭐

```python
pitch_value = 200 # adjuect value

high_pitched_audio  = augment.EffectChain().pitch(pitch_value).rate(rate).apply(audio_array, src_info={'rate': rate})
ipd.Audio(high_pitched_audio, rate=rate)
```

## Reverbration ⭐⭐⭐

```python
reverbered_audio = augment.EffectChain().reverb(100, 80, 90).channels(1).apply(audio_array, src_info={'rate': rate})
ipd.Audio(reverbered_audio, rate=rate)
```
